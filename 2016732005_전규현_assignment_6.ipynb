행렬 방정식 풀기(Pytorch 사용)  
---
* 다음 행렬 방정식을 'Pseudo inverse matrix'를 이용해 풀어보자.
* $A^TA$의 역행렬이 존재한다고 가정


>>>$Ax=B$  
$A=\begin{bmatrix}
0 & 1 \\
1 & 1 \\
2 & 1 \\
3 & 1 \end{bmatrix}$, $B=\begin{bmatrix}
-1 \\
0.2 \\
0.9 \\
2.1 \\ \end{bmatrix}$










```python
# 라이브러리에서 가져오기
import torch
```

```python
# 행렬 나타내기
A = torch.FloatTensor([[0,1],
                       [1,1],
                       [2,1],
                       [3,1]])
B = torch.FloatTensor([-1,0.2,0.9,2.1])

# A의 역행렬 구하기
inv_A = torch.linalg.pinv(A)

# A의 역행렬과 B를 곱하여 x의 값 구하기
x = inv_A.matmul(B)

# print함수로 나타내어주기
print("A=", A)
print("B=", B)
print("A^-1=", inv_A)
print("x=", x)
```

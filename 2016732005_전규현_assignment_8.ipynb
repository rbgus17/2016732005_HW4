### Assignment 8 
* Data 파일을 이용한 Logistic Regression
  - data_logistic_regression.csv을 이용해 학습한다.
    + 8개의 input, 1개의 output
>>> $[x_1,..x_8,y]$
* Accuracy 75% 이상을 달성한다.

```python
# 라이브러리 가져오기
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as func
import torch.optim as opt

# 구글 드라이브 마운트
from google.colab import drive
drive.mount('/content/drive')

# Seed 고정
torch.manual_seed(1)
```

```python
# Numpy를 이용해 csv file을 불러오기
dataset = np.loadtxt(
    '/content/drive/MyDrive/data_logistic_regression.csv', 
    delimiter=',', 
    dtype=np.float32) # 데이터 타입을 float32로 설정

# 순서를 random으로 섞기
np.random.shuffle(dataset)

# torch tensor로 변환
x_train = torch.FloatTensor(dataset[:,:-1])
y_train = torch.FloatTensor(dataset[:,[-1]])

# Model 설계
class Logistic(nn.Module):
    def __init__(self, num_inputs, num_outputs): # 초기화
        super().__init__()
        self.linear = nn.Linear(num_inputs, num_outputs) # linear model
        self.sigmoid = nn.Sigmoid() # sigmoid model

    def forward(self, x): # 입출력 텐서를 반환
        return self.sigmoid(self.linear(x)) # sigmoid로 linear입력 값 반환

# Model 초기화 (입력 dim=8, 출력 dim=1)
model = Logistic(8, 1)

# Optimizer 설정 (learning rate = 1.5로 설정)
optimizer = opt.SGD(model.parameters(), lr=1.5)

# 반복
for epoch in range(100):

  # Cost 계산 / mse_loss(가정에의한값, 참값)
  y_hypo = model(x_train)
  cost = func.binary_cross_entropy(y_hypo, y_train) # binary cross entropy 함수로 cost선언

  # cost를 이용해 model update
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  # 10번 마다 epoch, cost, accuracy 출력
  if epoch % 10 == 9:
    prediction = y_hypo >= torch.FloatTensor([0.5])
    correct_prediction = prediction.float() == y_train
    accuracy = correct_prediction.sum().item() / len(correct_prediction)
    print('Epoch {:4d}/{} Cost: {:.6f} Accuracy {:2.2f}%'.format(
            epoch, 100, cost.item(), accuracy * 100,
    ))
```

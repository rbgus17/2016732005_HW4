## 집합

```python
s1 = {1,3,2} # set은 순서가 없다.
s1
```

중복허용하지 않음.
순서가 없음. 따라서 list나 tuple로 전환해야 함

```python
s1 = {2,1,2,3,2} # set은 중복을 허용하지 않는다.
s1
```

교집합, 합집합, 차집합

```python
s1 = set([1,2,3,4])
s2 = set([1,2,5,6])
```

교집합

```python
s1.intersection(s2) # 교집합
```

```python
s1 & s2
```

합집합

```python
s1. union(s2)
```

```python
s1 | s2
```

차집합

```python
s1. difference(s2) # s1 중에서 s2와 다른 요소
```

```python
s1 - s2
```

```python
s2. difference(s1)
```

```python
s2 - s1
```

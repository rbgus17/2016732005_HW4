```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
# 그래프 초기화
plt.clf()
plt.plot(x, y_sin)
plt.plot(x, y_cos)
# 축이름 표시
plt.xlabel('x axis')
plt.ylabel('y axis')
# 제목 표시
plt.title('Sine and Cosine')
# 범주 표시 - 리스트를 이용
# 리스트 크기는 그래프 개수와 같아야 함
plt.legend(['Sine', 'Cosine'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.clf()
# 그래프 간격 조절
left=0.0
bottom=0.0
right=1.0
top=1.0
wspace=10.0
hspace=10.0
plt.subplots_adjust(left, bottom, right, top, wspace, hspace)
# 세로로 2개, 가로로 1개의 그래프 배치 선언
# 그중 1번에 그래프 생성
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine')
plt.xlabel('x axis')
plt.ylabel('y axis')
# 2번에 그래프 생성
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 그래프로 그릴 점들의 x좌표 값들과 y좌표 값들을 변수에 넣는다.
x = np.arange(0,10)
y = x
# plot함수를 이용해 그래프 그리기
plt.subplot(2,2,1)
plt.plot(x,x)
# 그래프 표시

# 그래프로 그릴 점들의 x좌표 값들과 y좌표 값들을 변수에 넣는다.
x = np.arange(-10,10)
y = x**2
# plot함수를 이용해 그래프 그리기
plt.subplot(2,2,2)
plt.plot(x,x**2)
# 그래프 표시

# 그래프로 그릴 점들의 
# x좌표 값들과 y좌표 값들을 변수에 넣는다
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
# plot함수를 이용해 점들을 그래프로 그린다
plt.subplot(2,2,3)
plt.plot(x, y)
# 그려진 그래프를 화면에 표시

# 그래프로 그릴 점들의 
# x좌표 값들과 y좌표 값들을 변수에 넣는다
x = np.arange(0, 3 * np.pi, 0.1)
y = np.cos(x)
# plot함수를 이용해 점들을 그래프로 그린다
plt.subplot(2,2,4)
plt.plot(x, y)
# 그려진 그래프를 화면에 표시
plt.show()
plt.clf()
# 그래프 간격 조절
left=0.0
bottom=0.0
right=1.0
top=1.0
wspace=1.0
hspace=1.0
plt.subplots_adjust(left, bottom, right, top, wspace, hspace)
```

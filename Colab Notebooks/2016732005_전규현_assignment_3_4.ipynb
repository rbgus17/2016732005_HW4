# numpy 실습
행렬 방정식 풀기

```python
# 라이브러리에서 가져오기
import numpy as np 
import numpy.linalg as lin
# 행렬 나타내기
matA = np.array([[0,1],[1,1],[2,1],[3,1]])
matB = np.array([[-1.0],[0.2],[0.9],[2.1]])

print('A=',matA)
print('B=',matB)
# A의 역행렬 구하기
inv_a = lin.pinv(matA)
# A의 역행렬과 B를 곱하여 x의 값 구하기
x = np.dot(inv_a, matB)
print('A^-1=',inv_a)
print('x=',x)
```

# matplotlib 실습
2x2 그래프 그리기

```python
# 라이브러리에서 가져오기
import numpy as np
import matplotlib.pyplot as plt

plt.clf()
# 그래프 간격 조절
left=0.0
bottom=0.0
right=1.0
top=1.0
wspace=0.5
hspace=0.5
plt.subplots_adjust(left, bottom, right, top, wspace, hspace)

# y=x 
# 그래프로 그릴 점들의 x좌표 값들과 y좌표 값들을 변수에 넣는다.
x = np.arange(0,10)
y = x
# plot함수를 이용해 그래프 그리기
plt.subplot(2,2,1)
plt.plot(x,x)
# 축이름 표시
plt.xlabel('x axis')
plt.ylabel('y axis')
# 범주 표시
plt.legend(['$y=x$'])

# y=x^2
# 그래프로 그릴 점들의 x좌표 값들과 y좌표 값들을 변수에 넣는다.
x = np.arange(-10,10)
y = x**2
# plot함수를 이용해 그래프 그리기
plt.subplot(2,2,2)
plt.plot(x,x**2)
# 축이름 표시
plt.xlabel('x axis')
plt.ylabel('y axis')
# 범주 표시
plt.legend(['$y=x^2$'])

# y=sin(x)
# 그래프로 그릴 점들의 
# x좌표 값들과 y좌표 값들을 변수에 넣는다
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
# plot함수를 이용해 점들을 그래프로 그린다
plt.subplot(2,2,3)
plt.plot(x, y)
# 축이름 표시
plt.xlabel('x axis')
plt.ylabel('y axis')
# 범주 표시
plt.legend(['$y=sin(x)$'])

# y=cos(x)
# 그래프로 그릴 점들의 
# x좌표 값들과 y좌표 값들을 변수에 넣는다
x = np.arange(0, 3 * np.pi, 0.1)
y = np.cos(x)
# plot함수를 이용해 점들을 그래프로 그린다
plt.subplot(2,2,4)
plt.plot(x, y)
# 축이름 표시
plt.xlabel('x axis')
plt.ylabel('y axis')
# 범주 표시
plt.legend(['$y=cos(x)$'])
# 그려진 그래프를 화면에 표시
plt.show()

```
